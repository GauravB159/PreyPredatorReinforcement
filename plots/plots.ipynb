```python
import matplotlib.pyplot as plt
import pandas
```

```python
def line_plot(title, xlabel, ylabel, x, y1, y2=None, y3=None, y4=None, y5=None, y6=None):
    
    legend = [y1.name]

    plt.plot(x, y1)
    if y2 is not None:
        plt.plot(x, y2)
        legend.append(y2.name)
    if y3 is not None:
        plt.plot(x, y3)
        legend.append(y3.name)
    if y4 is not None:
        plt.plot(x, y4)
        legend.append(y4.name)
    if y5 is not None:
        plt.plot(x, y5)
        legend.append(y5.name)
    if y6 is not None:
        plt.plot(x, y6)
        legend.append(y6.name)

    if y2 is not None:
        plt.legend(legend)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
def scatter_plot(title, xlabel, ylabel, x, y1, y2=None, y3=None, y4=None, y5=None, y6=None):
    
    legend = [y1.name]

    plt.scatter(x, y1, alpha=0.5)
    if y2 is not None:
        plt.scatter(x, y2, alpha=0.5)
        legend.append(y2.name)
    if y3 is not None:
        plt.scatter(x, y3, alpha=0.5)
        legend.append(y3.name)
    if y4 is not None:
        plt.scatter(x, y4, alpha=0.5)
        legend.append(y4.name)
    if y5 is not None:
        plt.scatter(x, y5, alpha=0.5)
        legend.append(y5.name)
    if y6 is not None:
        plt.scatter(x, y6, alpha=0.5)
        legend.append(y6.name)

    if y2 is not None:
        plt.legend(legend)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

# One Prey

```python
one_prey = pandas.read_csv("ppo_log.csv")
```

```python
line_plot('Single Prey Agent (40 x 40 Grid): Reward by Episode', 'Episode', 'Reward', 
          one_prey['Episode'], one_prey['Avg reward'])
```

```python
scatter_plot('Single Prey Agent (40 x 40 Grid): Reward by Episode Length', 'Episode Length', 'Reward', 
             one_prey['Avg length'], one_prey['Avg reward'])
```

# Predator and Prey: MAPPO

```python
pp = pandas.read_csv('ppo_log_ma.csv')
pp['Total Average reward'] = pp['Avg prey reward'] + pp['Avg predator reward']
pp.head()
```

```python
pp.columns = ['Episode', 'Average Length', 'Average Prey Reward', 'Average Predator Reward', 'Total Average Reward']
```

```python
line_plot('Average Predator and Prey Rewards By Episode', 'Episode', 'Reward', 
          pp['Episode'], pp['Total Average Reward'], pp['Average Prey Reward'], pp['Average Predator Reward'])
```

```python
scatter_plot('Average Predator and Prey Rewards By Episode', 'Average Episode Length', 'Reward', 
             pp['Average Length'], pp['Total Average Reward'], pp['Average Prey Reward'], pp['Average Predator Reward'])
```

#  Predator and Prey: Ping Pong

```python
pp2 = pandas.read_csv('ppo_log_ma_2.csv')
pp2['Avg total reward'] = pp2['Avg prey reward'] + pp2['Avg predator reward']
pp2.head()
```

```python
pp2.columns = ['Episode', 'Average Length', 'Total Prey Reward', 'Average Prey Reward', 
               'Total Predator Reward', 'Average Predator Reward', 'Average Total Reward']
pp2['Total Reward'] = pp2['Total Prey Reward'] + pp2['Total Predator Reward']
pp2.head()
```

```python
line_plot('Average Predator and Prey Rewards (Ping Pong) By Episode', 'Episode', 'Reward', 
          pp2['Episode'], pp2['Average Total Reward'], pp2['Average Prey Reward'], pp2['Average Predator Reward'])
```

```python
line_plot('Total Predator and Prey Rewards (Ping Pong) By Episode', 'Episode', 'Reward', 
          pp2['Episode'], pp2['Total Reward'], pp2['Total Prey Reward'], pp2['Total Predator Reward'])
```

```python
scatter_plot('Average Predator and Prey Rewards (Ping Pong) By Average Episode Length', 'Average Episode Length', 'Reward', 
          pp2['Average Length'], pp2['Average Total Reward'], pp2['Average Prey Reward'], pp2['Average Predator Reward'])
```

```python
scatter_plot('Total Predator and Prey Rewards (Ping Pong) By Average Episode Length', 'Average Episode Length', 'Reward', 
          pp2['Average Length'], pp2['Total Reward'], pp2['Total Prey Reward'], pp2['Total Predator Reward'])
```
