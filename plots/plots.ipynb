```python
import matplotlib.pyplot as plt
import pandas
```

```python
def line_plot(x, y, title, xlabel, ylabel):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()


def line_plot_multiple(x, y1, y2, y3, title, xlabel, ylabel):
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.plot(x, y3)
    plt.legend([y1.name, y2.name, y3.name])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
def scatter_plot(x, y, title, xlabel, ylabel):
    plt.scatter(x, y, alpha=0.5)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

def scatter_plot_multiple(x, y1, y2, y3, title, xlabel, ylabel):
    plt.scatter(x, y1, alpha=0.5)
    plt.scatter(x, y2, alpha=0.5)
    plt.scatter(x, y3, alpha=0.5)
    plt.legend([y1.name, y2.name, y3.name])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

# One Prey

```python
one_prey = pandas.read_csv("ppo_log.csv")
```

```python
line_plot(one_prey['Episode'], one_prey['Avg reward'], 'Single Prey Agent (40 x 40 Grid): Reward by Episode', 'Episode', 'Reward')
```

```python
scatter_plot(one_prey['Avg length'], one_prey['Avg reward'], 'Single Prey Agent (40 x 40 Grid): Reward by Episode Length', 'Episode Length', 'Reward')
```

# Predator and Prey

```python
pp = pandas.read_csv('ppo_log_ma.csv')
pp['Avg total reward'] = pp['Avg prey reward'] + pp['Avg predator reward']
pp.head()
```

```python
pp.columns = ['Episode', 'Average Length', 'Average Prey Reward', 'Average Predator Reward', 'Average Total Reward']
```

```python
line_plot_multiple(pp['Episode'], pp['Average Total Reward'], pp['Average Prey Reward'], pp['Average Predator Reward'], 'Average Predator and Prey Rewards By Episode', 'Episode', 'Reward')
```

```python
scatter_plot_multiple(pp['Average Length'], pp['Average Total Reward'], pp['Average Prey Reward'], pp['Average Predator Reward'], 'Average Predator and Prey Rewards By Episode', 'Average Episode Length', 'Reward')
```
