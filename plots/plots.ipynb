```python
import matplotlib.pyplot as plt
import pandas
```

```python
def line_plot(x, y, title, xlabel, ylabel):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
def scatter_plot(x, y, title, xlabel, ylabel):
    plt.scatter(x, y, alpha=0.5)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

# One Prey

```python
one_prey = pandas.read_csv("ppo_log.csv")
```

```python
line_plot(one_prey['Episode'], one_prey['Avg reward'], 'Single Prey Agent (40 x 40 Grid): Reward by Episode', 'Episode', 'Reward')
```

```python
scatter_plot(one_prey['Avg length'], one_prey['Avg reward'], 'Single Prey Agent (40 x 40 Grid): Reward by Episode Length', 'Episode Length', 'Reward')
```
